```python
ct_path = './data/ct_images/full_archive/'
data = np.load(ct_path + 'image.npy', allow_pickle=True)
len(data)
```

```python
img1 = data[0]
plt.imshow(img1, cmap='gray')
plt.show()
```

```python
print('max', img1.max(), 'min', img1.min())
```

```python
img1_normalize = ((img1 - img1.min())/(img1.max() - img1.min()))*255
print('max', img1_normalize.max(), 'min', img1_normalize.min())
```

```python
plt.imshow(img1_normalize, cmap='gray')
plt.show()
```

```python
img_exp = np.expand_dims(img1_normalize, axis=-1)
img_cat = np.concatenate([img_exp, img_exp, img_exp], axis=-1)
img_cat = img_cat.astype('uint8')
img_cat.shape
```

```python
print('max', img_cat.max(), 'min', img_cat.min())
print(img_cat.dtype)
```

```python
for i in range(len(data)):
    if len(data[i].shape) == 2:
        img_sub = data[i]
        img_normalize = ((img_sub - img_sub.min())/(img_sub.max() - img_sub.min()))*255
        img_exp = np.expand_dims(img_normalize, axis=-1)
        img_cat = np.concatenate([img_exp, img_exp, img_exp], axis=-1)
        img_cat = img_cat.astype('uint8')
        skimage.io.imsave(ct_path + '../images/' + str(i+1) + '.jpg', img_cat)
```
